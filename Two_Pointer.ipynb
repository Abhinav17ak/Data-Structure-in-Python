```python
def two_sum(arr , tar):
    left = 0
    right = len(arr) - 1
    while left < right :
        if arr[left] + arr[right] == tar:
            return (left , right)
        elif arr[left] + arr[right] > tar:
            right -= 1
        else :
            left += 1
    return -1 , -1
arr = [20 , 40 , 60 , 80 , 90 , 120 , 240 ]
target_sum = 120
res = two_sum(arr , target_sum)
print(res)

```
