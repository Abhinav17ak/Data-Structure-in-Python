# Binary Search

```python
def BinarySearch(arr , x , i , j ):
    while i <= j :
        mid = i+ (j-i)//2
        if arr[mid] == x :
            return mid
        elif arr[mid] < x :
            return BinarySearch(arr , x , mid+1 , j)
        else:
            
            return BinarySearch(arr , x , i , mid-1)
    return -1


#  Driver Code
arr = [1, 5 , 10 , 14 , 20 , 58 , 90 , 100]
x = 90
i = 0
j = len(arr) - 1
result = BinarySearch(arr , x , i , j)
print("Search element Presenty at the location  ", result)
    
```
